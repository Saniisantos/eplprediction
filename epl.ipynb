```python
def rolling_averages(df, cols, new_cols):
    df_copy = df.copy()
    # Apply rolling average logic here
    for col in cols:
        # Example for calculating rolling average
        df_copy[f'{col}_rolling'] = df_copy[col].rolling(window=5).mean()
    
    # Add new columns for the rolling averages
    for col, new_col in zip(cols, new_cols):
        df_copy[new_col] = df_copy[f'{col}_rolling']
    
    # Ensure team is kept
    df_copy['team'] = df['team']
    
    return df_copy
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < "2022-01-01"].copy()  # Copy to avoid warning
    test = data[data["date"] > "2022-01-01"].copy()  # Copy to avoid warning

    print("Train columns:", train.columns)  # Debugging step
    print("Predictors being used:", predictors)

    # Check for missing columns
    missing_columns = [col for col in predictors if col not in train.columns]
    if missing_columns:
        raise KeyError(f"Missing columns in train data: {missing_columns}")

    # Fill NaN values correctly
    train.loc[:, predictors] = train[predictors].fillna(0)
    test.loc[:, predictors] = test[predictors].fillna(0)

    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])

    combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
    return combined, (preds == test["target"]).mean()
```
